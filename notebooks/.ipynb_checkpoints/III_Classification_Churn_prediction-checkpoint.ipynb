## Feature Engineering: Mutual Information

Technique to realise the importance of categorical features

```python
from sklearn.metrics import mutual_info_score
```

```python
mutual_info_score(df_full_train.contract, df_full_train.churn)
```

```python
mutual_info_score(df_full_train.gender, df_full_train.churn)
```

```python
mutual_info_score(df_full_train.partner, df_full_train.churn)
```

```python
def mutual_info_churn_score(series):
    return mutual_info_score(series, df_full_train["churn"])
    
```

```python
mi = df_full_train[categorical].apply(mutual_info_churn_score)
mi.sort_values(ascending=False)
```

```python
categorical.remove("churn")
train_dicts = df_train[categorical+numerical].to_dict(orient="records")
dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)
```

```python
val_dicts = df_val[categorical+numerical].to_dict(orient="records")
X_val = dv.fit_transform(val_dicts)
```

## EVALUATION METRICS

###  ACCURACY

## Accuracy = (Number of Cprrect Predictions) / (Total Number of preditions)

But Accuracy has a drawback -> In our baseline model,even though which predicts 0 all the time, has high accuracy better than our model.
So,accuracy doesnt tell good metric if distribution of y is skewed i.e. more than 70% are 0s and rest are 1s i.e. In Class Imbalance problems accuracy can be misleading.

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_val, y_pred)
```

### CONFUSION MATRIX

TN        FP   

FN        TP

## ACCURACY

(tp + tn)/(tp + tn + fn + fp)

## PRECISION

It tell us fractions of positive predictions are actually correct.

(tp) / (tp + fp)

## RECALL

Fraction of correctly idenitifed positive examples

(tp)/ (tp + fn)

## ROC CURVE

Plot TPR (True Positive rate) on Y-axis and FPR(False positive rate) on x-axis.

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_val, y_pred)
```

```python
plt.plot(fpr, tpr)
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("FPR")
plt.ylabel("TPR")

plt.legeend()
```

## ROC-AUC

Receiver Operating Characteristic - Area Under the Curve

The ideal point is when FPR=0 and TPR=1 i,e. top opint on y axis, So , roc must be close to this point. To quantify this, the area under the curve should be as much as possible

```python
from sklearn.metrics import auc
```

```python
auc(fpr, tpr)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_val, y_pred)
```
