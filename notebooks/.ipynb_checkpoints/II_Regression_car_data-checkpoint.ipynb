#### Linear Regression

```python
def dot(xi, w):
    n = len(xi)
    res = 0.0

    for j in range(n):
        res = res + w[j]*xi[j]
    
    return res


def linear_regression(xi, w):
    xi = [1] + xi
    w_new = [w0] + w
    return dot(xi, w)
```

```python
def train_linear_regression(X, y):
    
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
     
    term1 = np.linalg.inv(X.T.dot(X))
    result = term1.dot(X.T).dot(y)
    
    return result[0], result[1:]
    
    
```

```python
def calculate_rmse(y_train, y_pred):
    diff_squared = np.mean((y_train-y_pred)**2)
    return np.sqrt(diff_squared)

calculate_rmse(y_train, y_pred)
```

#### Validating the model

```python
def prepare_X(df):
    df_num = df[base]
    return df_num[base].fillna(0).values

X_train = prepare_X(df_train)
w0, w = train_linear_regression(X_train, y_train)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

calculate_rmse(y_pred, y_val)
```

#### Regularization

```python
XTX = np.array([[1,2,2 ],[2,1,1 ],[2,1,1 ]])
```

```python
np.linalg.inv(XTX)
```

For the above matrix, the inverse cannot be computed and so we cannot the coeeficients. To solve this problem, we add a small number 
to diagonals of the matrix to regularize the weights i.e. bring them under control.

```python
XTX = np.array([[1.0001,2,2 ],[2,1.01,1 ],[2,1,1.000000003 ]])
np.linalg.inv(XTX)
```

```python
def train_linear_regression_reg(X, y, r=0.01):
    
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    
    M = X.T.dot(X) + r*np.eye(X.T.dot(X).shape[0]) 
    term1 = np.linalg.inv(M)
    result = term1.dot(X.T).dot(y)
    
    return result[0], result[1:]
```

```python
X_train = prepare_X(df_train)
w0, w = train_linear_regression_reg(X_train, y_train,r=0.00001)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

calculate_rmse(y_pred, y_val)
```

#### Fine tuning the model

```python
for r in [0, 0.01, 0.0001, 0.000000001, 1000]:
    X_train = prepare_X(df_train)
    w0, w = train_linear_regression_reg(X_train, y_train,r=0.00001)
    
    X_val = prepare_X(df_val)
    y_pred = w0 + X_val.dot(w)
    
    score = calculate_rmse(y_pred, y_val)
    print(r, w0, score)
```
