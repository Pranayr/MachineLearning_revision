```python
## Split the data set into - train and test data set.
## Create a linear regression model with all features and remove all NaNs.
## Feature engineering and then linear regression model, filling in NaNs.
## Remvoe unwanted features and then a linear regression model again.
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
```

```python
data.dropna(inplace=True)
```

```python
numerical = list(set(data.columns) - set(categorical))
```

```python
numerical.remove("msrp")
```

```python
numerical.remove("year")
numerical
```

```python
X = data[categorical + numerical]
y = data["msrp"]      

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=27
)
```

```python
X_train.shape, X_test.shape
```

```python
X_train_he = ohe.fit_transform(X_train)
X_test_he = ohe.transform(X_test)
```

```python
X_train_he.shape, X_test_he.shape
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()

model.fit(X_train_he, y_train)

y_pred = model.predict(X_test_he)
```

```python
print("Intercept (0):", model.intercept_)
print("Coefficients (1, 2,...):", model.coef_)

print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
print("R Score:", r2_score(y_test, y_pred))
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge, Lasso

ridge = Ridge(alpha=1.0)  # alpha = regularization strength
ridge.fit(X_train_he, y_train)
y_pred_ridge = ridge.predict(X_test_he)

print("Ridge Coefficients:", ridge.coef_)
print("Ridge Intercept:", ridge.intercept_)
print("Ridge MSE:", mean_squared_error(y_test, y_pred_ridge))
print("Ridge R2:", r2_score(y_test, y_pred_ridge))

```

## Lasso Regression

```python
# ---------------------
# Lasso Regression
lasso = Lasso(alpha=0.1)  # smaller alpha -> less regularization
lasso.fit(X_train_he, y_train)
y_pred_lasso = lasso.predict(X_test_he)

print("Lasso Coefficients:", lasso.coef_)
print("Lasso Intercept:", lasso.intercept_)
print("Lasso MSE:", mean_squared_error(y_test, y_pred_lasso))
print("Lasso R2:", r2_score(y_test, y_pred_lasso))
```
